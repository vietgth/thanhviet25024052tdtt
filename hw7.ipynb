## W7A16 
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
a = list(map(int, input().split()))
dem = {}
i = 0
while i < len(a):
    if a[i] not in dem:
        dem[a[i]] = 0
    dem[a[i]] += 1
    i += 1

# Xóa phần tử nào không thỏa: giữ lại min(x, dem[x]) bản của x
result = []
for x in dem:
    if x == 0:
        continue  # 0 phải xuất hiện 0 lần -> xóa hết
    giu = x if dem[x] >= x else 0  # chỉ giữ nếu đủ x bản
    j = 0
    while j < giu:
        result.append(x)
        j += 1

print(result)
```

## W7A17 
Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì 

```python
a = list(map(int, input().split()))
candidate = None
count = 0
i = 0
while i < len(a):
    if count == 0:
        candidate = a[i]
        count = 1
    elif a[i] == candidate:
        count += 1
    else:
        count -= 1
    i += 1

count = 0
i = 0
while i < len(a):
    if a[i] == candidate:
        count += 1
    i += 1

if count > len(a) // 2:
    print(candidate)
```

## W7A18 
Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng 
dần, lẻ giảm dần; ;  

```python
a = list(map(int, input().split()))
chan = []
le = []
i = 0
while i < len(a):
    if a[i] % 2 == 0:
        chan.append(a[i])
    else:
        le.append(a[i])
    i += 1

# bubble sort tăng cho chẵn
i = 0
while i < len(chan):
    j = 0
    while j < len(chan) - i - 1:
        if chan[j] > chan[j+1]:
            chan[j], chan[j+1] = chan[j+1], chan[j]
        j += 1
    i += 1

# bubble sort giảm cho lẻ
i = 0
while i < len(le):
    j = 0
    while j < len(le) - i - 1:
        if le[j] < le[j+1]:
            le[j], le[j+1] = le[j+1], le[j]
        j += 1
    i += 1

print(chan + le)
```

## W7A19 
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không 
dùng bất cứ thuật toán sắp xếp nào

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))

# Xác định chiều tăng/giảm
tang_a = len(a) < 2 or a[0] <= a[-1]
tang_b = len(b) < 2 or b[0] <= b[-1]

result = []
i = j = 0
if tang_a and tang_b:
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            result.append(a[i]); i += 1
        else:
            result.append(b[j]); j += 1
    while i < len(a): result.append(a[i]); i += 1
    while j < len(b): result.append(b[j]); j += 1
else:
    while i < len(a) and j < len(b):
        if a[i] >= b[j]:
            result.append(a[i]); i += 1
        else:
            result.append(b[j]); j += 1
    while i < len(a): result.append(a[i]); i += 1
    while j < len(b): result.append(b[j]); j += 1

print(result)
```

## W7A22 
Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. 
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
n = int(input())
dem = {}
i = 0
while i < n:
    ho_ten = input().split()
    ten = ho_ten[-1]
    if ten not in dem:
        dem[ten] = 0
    dem[ten] += 1
    i += 1

max_so_lan = 0
for ten in dem:
    if dem[ten] > max_so_lan:
        max_so_lan = dem[ten]

ket_qua = None
for ten in dem:
    if dem[ten] == max_so_lan:
        if ket_qua is None or ten < ket_qua:
            ket_qua = ten

print(ket_qua)
```

## W7A25 
Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi 
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số 

```python
s = input()
k = int(input())
n = len(s)
result = ""
start = 0
i = 0
while i < k:
    con_lai = k - i
    max_pos = start
    j = start
    while j <= n - con_lai:
        if s[j] > s[max_pos]:
            max_pos = j
        j += 1
    result += s[max_pos]
    start = max_pos + 1
    i += 1
print(result)
```

## W7A26 
Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các 
số đó.  

```python
s = input()
numbers = []
i = 0
while i < len(s):
    if s[i] >= '0' and s[i] <= '9':
        num = 0
        while i < len(s) and s[i] >= '0' and s[i] <= '9':
            num = num * 10 + int(s[i])
            i += 1
        numbers.append(num)
    else:
        i += 1

if len(numbers) == 0:
    print("Khong co so")
else:
    tong = 0
    i = 0
    while i < len(numbers):
        tong += numbers[i]
        i += 1
    print("Tong:", tong)
    print("Trung binh:", tong / len(numbers))
```

## W7A27 
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ 
hay không 

```python
s = input()
stack = []
hop_le = 1
dong = {')': '(', '}': '{', ']': '['}
i = 0
while i < len(s):
    c = s[i]
    if c in '({[':
        stack.append(c)
    elif c in ')}]':
        if len(stack) == 0 or stack[-1] != dong[c]:
            hop_le = 0
        else:
            stack.pop()
    i += 1
if len(stack) != 0:
    hop_le = 0
print("Hop le" if hop_le else "Khong hop le")
```

## W7A28 
Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên 
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ 
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua) 

```python
def la_hop_so(x):
    if x < 4:
        return 0
    i = 2
    while i * i <= x:
        if x % i == 0:
            return 1
        i += 1
    return 0

n = int(input())
best_start = -1
best_len = 0
cur_start = -1
cur_len = 0
x = n + 1
while x <= 10000:  # giới hạn tìm kiếm
    if la_hop_so(x):
        if cur_len == 0:
            cur_start = x
        cur_len += 1
        if cur_len > best_len:
            best_len = cur_len
            best_start = cur_start
    else:
        cur_len = 0
    x += 1

if best_len == 0:
    print("Khong co")
else:
    i = best_start
    while i < best_start + best_len:
        print(i, end=" ")
        i += 1
    print()
```

## W7A29 
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau 

```python
a = list(map(int, input().split()))
best_start = 0
best_len = 1
cur_start = 0
cur_len = 1
i = 1
while i < len(a):
    if a[i] > a[i-1]:
        cur_len += 1
        if cur_len > best_len:
            best_len = cur_len
            best_start = cur_start
    else:
        cur_start = i
        cur_len = 1
    i += 1
print(a[best_start:best_start + best_len])
```

## W7A30 
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím 

```python
a = list(map(int, input().split()))
n = len(a)
dp = [1] * n
prev = [-1] * n

i = 1
while i < n:
    j = 0
    while j < i:
        if a[j] < a[i] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1
            prev[i] = j
        j += 1
    i += 1

max_len = 0
end_idx = 0
i = 0
while i < n:
    if dp[i] > max_len:
        max_len = dp[i]
        end_idx = i
    i += 1

result = []
idx = end_idx
while idx != -1:
    result.append(a[idx])
    idx = prev[idx]

result.reverse()
print(result)
```
