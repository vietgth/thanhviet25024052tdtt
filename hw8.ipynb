Xây dựng lớp Shape với phương thức area(). 
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area(). 
Lấy π = 3.14. 
 
Input 
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE) 
Dòng 2: tham số tương ứng. 
 
Output 
In ra diện tích với 2 chữ số thập phân.

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r

loai = input()
if loai == "RECTANGLE":
    w, h = map(float, input().split())
    print("%.2f" % Rectangle(w, h).area())
else:
    r = float(input())
    print("%.2f" % Circle(r).area())
```

Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance(). 
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức 
add_interest() để cộng balance*rate vào số dư. 
 
Input 
balance rate 
 
Output 
Số dư sau khi cộng lãi, in 2 chữ số thập phân.

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def get_balance(self):
        return self.balance

class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate
    def add_interest(self):
        self.balance += self.balance * self.rate

balance, rate = map(float, input().split())
acc = SavingAccount(balance, rate)
acc.add_interest()
print("%.2f" % acc.get_balance())
```

## W8A21
 Yêu cầu 

Xây dựng lớp Printer với phương thức print_value(x). 
Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của 
số nguyên dương x. 
 
Input 
Một số nguyên dương n. 
 
Output 
In ra dạng nhị phân của n. 

```python
class Printer:
    def print_value(self, x):
        print(x)

class BinaryPrinter(Printer):
    def print_value(self, x):
        if x == 0:
            print(0)
            return
        result = ""
        while x > 0:
            result = str(x % 2) + result
            x //= 2
        print(result)

n = int(input())
BinaryPrinter().print_value(n)
```

Yêu cầu 

 Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận: 
[a b] 
[c d] 
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc. 
 
Input 
Một dòng chứa 4 số nguyên a b c d. 
 
Output 
In ra trace và determinant trên cùng một dòng.

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a * self.d - self.b * self.c

a, b, c, d = map(int, input().split())
m = Matrix2x2(a, b, c, d)
print(m.trace(), m.determinant())
```

Xây dựng lớp Product(name, price). 
Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức 
final_price() để trả về giá sau giảm. 
Công thức: price * (100 - percent) / 100 
 
Input 
name price percent 
 
Output 
In ra giá cuối cùng với 2 chữ số thập phân.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def final_price(self):
        return self.price

class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent
    def final_price(self):
        return self.price * (100 - self.percent) / 100

data = input().split()
name = data[0]
price = float(data[1])
percent = float(data[2])
print("%.2f" % DiscountProduct(name, price, percent).final_price())
```

 Yêu cầu 

Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), 
value(). 
Sau đó thực hiện dãy thao tác trên đối tượng. 
 
Input 
Dòng 1: n 
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách. 
 
Output 
Giá trị cuối cùng của count.

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count

n = int(input())
ops = input().split()
c = Counter()
i = 0
while i < n:
    if ops[i] == "INC":
        c.inc()
    else:
        c.dec()
    i += 1
print(c.value())
```

Xây dựng lớp Time(h, m, s) với phương thức to_seconds() đổi thời gian sang tổng số giây.

Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), 
pop(), top(). 
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ. 
 
Input 
Dòng 1: n 
n dòng tiếp theo là một trong ba lệnh: 
PUSH x 
POP 
TOP 
 
Output 
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack:
    def __init__(self):
        self.data = []
    def push(self, x):
        self.data.append(x)
    def pop(self):
        self.data.pop()
    def top(self):
        return self.data[-1]

n = int(input())
st = Stack()
i = 0
while i < n:
    cmd = input().split()
    if cmd[0] == "PUSH":
        st.push(int(cmd[1]))
    elif cmd[0] == "POP":
        st.pop()
    else:
        print(st.top())
    i += 1
```
