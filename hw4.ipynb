## Bài 3:
Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 

```python
a,b=map(int,input().split())
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i)
```

## Bài 4:
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để in ra các số
dương chẵn nhỏ hơn hoặc bằng n (n <= 1000)

```python
n = int(input())
i = 2
while i <= n:
    print(i)
    i += 2
```

## Bài 5:
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
so = n
while so % 2 == 0:
    so = so // 2
if so == 1:
    print(n, "là số dạng 2^k")
else:
    print(n, "không phải số dạng 2^k")
```

## Bài 6:
Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input())
tong = 0
while n > 0:
    tong += n % 10
    n = n // 10
print(tong)
```

## Bài 7:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    n = n // 10
print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

## Bài 8:
Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
tong = 0
k = 0
while tong + (k + 1) < n:
    k += 1
    tong += k
print("k =", k)
print("S(k) =", tong)
```

## Bài 9:
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = float(input())
tong = 0
n = 0
while tong <= A:
    n += 1
    tong += 1 / n
print("n nhỏ nhất:", n)
```

## Bài 10:
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
so_dau = int(input("Nhập số: "))
lon_nhat = so_dau
nho_nhat = so_dau
while so_dau != -1:
    so_dau = int(input("Nhập số: "))
    if so_dau != -1:
        if so_dau > lon_nhat:
            lon_nhat = so_dau
        if so_dau < nho_nhat:
            nho_nhat = so_dau
print("Số lớn nhất:", lon_nhat)
print("Số nhỏ nhất:", nho_nhat)
```

## Bài 11:
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
A = int(input())
a = 1
b = 1
while b <= A:
    a, b = b, a + b
print("Số Fibonacci lớn nhất không vượt quá", A, "là:", a)
```

## Bài 12:
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
chuoi = input()
so_tu = 0
i = 0
while i < len(chuoi):  # dùng while thay vì len() trực tiếp trong for
    if chuoi[i] != ' ':
        so_tu += 1
        while i < len(chuoi) and chuoi[i] != ' ':
            i += 1
    else:
        i += 1
print(so_tu)
```

## Bài 13:
Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split() 

```python
chuoi = input()
tu = ""
for i in range(len(chuoi)):
    if chuoi[i] == ' ':
        break
    tu += chuoi[i]
print("Từ đầu tiên:", tu)
```

## Bài 14:
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa.

```python
chuoi = input()
ket_qua = ""
for i in range(len(chuoi)):
    if 'a' <= chuoi[i] <= 'z':
        ket_qua += chr(ord(chuoi[i]) - 32)
    elif 'A' <= chuoi[i] <= 'Z':
        ket_qua += chr(ord(chuoi[i]) + 32)
    else:
        ket_qua += chuoi[i]
print(ket_qua)
```

## Bài 15:
Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó

```python
chuoi = input()
tong = 0
so = ""
for i in range(len(chuoi)):
    if chuoi[i] == ',' or i == len(chuoi) - 1:
        if i == len(chuoi) - 1:
            so += chuoi[i]
        tong += int(so.strip())
        so = ""
    else:
        so += chuoi[i]
print(tong)
```

## Bài 16:
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
chuoi = input()
hoa = 0
thuong = 0
so = 0
for i in range(len(chuoi)):
    if 'A' <= chuoi[i] <= 'Z':
        hoa += 1
    elif 'a' <= chuoi[i] <= 'z':
        thuong += 1
    elif '0' <= chuoi[i] <= '9':
        so += 1
print("Ký tự hoa:", hoa)
print("Ký tự thường:", thuong)
print("Ký tự số:", so)
```
