## W9A16 
Cho hai số nguyên dương, n và m (0 < n, m <= 20 ). .Bạn hãy in ra tất cả các chuỗi nhị phân có 
n số 0 và m số 1. In theo thứ tự tăng dần trong từ điển 

```python
def sinh(current, so_0, so_1, n, m):
    if so_0 == n and so_1 == m:
        print(current)
        return
    if so_0 < n:
        sinh(current + "0", so_0 + 1, so_1, n, m)
    if so_1 < m:
        sinh(current + "1", so_0, so_1 + 1, n, m)

n, m = map(int, input().split())
sinh("", 0, 0, n, m)
```

## W9A17
 Viết hàm tìm dãy con tăng nghiêm ngặt dài nhất trong dãy đầu vào từ bàn phím 

```python
a = list(map(int, input().split()))
n = len(a)
dp = [1] * n
prev = [-1] * n

i = 1
while i < n:
    j = 0
    while j < i:
        if a[j] < a[i] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1
            prev[i] = j
        j += 1
    i += 1

max_len = 0
end_idx = 0
i = 0
while i < n:
    if dp[i] > max_len:
        max_len = dp[i]
        end_idx = i
    i += 1

result = []
idx = end_idx
while idx != -1:
    result.append(a[idx])
    idx = prev[idx]
result.reverse()
print(result)
```

## W9A18 
Cho dãy N phần tử số nguyên, hãy đếm số cặp có vị trí khác nhau và có tổng bằng k (số nguyên 
dương cho vào từ bàn phím)

```python
a = list(map(int, input().split()))
k = int(input())
dem = 0
i = 0
while i < len(a):
    j = i + 1
    while j < len(a):
        if a[i] + a[j] == k:
            dem += 1
        j += 1
    i += 1
print(dem)
```

## W9A20 
Thực hiện chuẩn hóa chuỗi kí tự: 
 
a. bỏ khoảng trắng đầu cuối 
 
b. các khoảng trắng giữa chỉ còn 1 dấu cách 
 
c. viết hoa chữ cái đầu mỗi từ 
KHÔNG DÙNG CÁC HÀM SẴN CỐ  

```python
s = input()

# Bỏ khoảng trắng đầu cuối
start = 0
while start < len(s) and s[start] == ' ':
    start += 1
end = len(s) - 1
while end >= 0 and s[end] == ' ':
    end -= 1
s = s[start:end+1]

# Rút gọn khoảng trắng giữa còn 1
i = 0
result = ""
prev_space = 0
while i < len(s):
    if s[i] == ' ':
        if prev_space == 0:
            result += ' '
        prev_space = 1
    else:
        result += s[i]
        prev_space = 0
    i += 1

# Viết hoa chữ cái đầu mỗi từ
words = result.split(' ')
final = ""
w = 0
while w < len(words):
    word = words[w]
    if len(word) > 0:
        first = word[0]
        if 'a' <= first <= 'z':
            first = chr(ord(first) - 32)
        new_word = first + word[1:]
    else:
        new_word = word
    if w > 0:
        final += ' '
    final += new_word
    w += 1

print(final)
```

## W9A21 
Tìm phần tử lớn thứ k trong dãy n phần tử bằng thuật toán tốt nhất có thể – Tránh dùng sắp xếp 
(dùng phân hoạch của quicksort) 

```python
def phan_hoach(a, lo, hi):
    pivot = a[hi]
    i = lo - 1
    j = lo
    while j < hi:
        if a[j] >= pivot:
            i += 1
            a[i], a[j] = a[j], a[i]
        j += 1
    a[i+1], a[hi] = a[hi], a[i+1]
    return i + 1

def quickselect(a, lo, hi, k):
    if lo == hi:
        return a[lo]
    p = phan_hoach(a, lo, hi)
    rank = p - lo + 1
    if k == rank:
        return a[p]
    elif k < rank:
        return quickselect(a, lo, p - 1, k)
    else:
        return quickselect(a, p + 1, hi, k - rank)

a = list(map(int, input().split()))
k = int(input())
print(quickselect(a, 0, len(a) - 1, k))
```

## W9A22 
Hai chuỗi là anagram nếu chúng có cùng ký tự và số lần xuất hiện. viết CT kiểm tra xem 2 chuỗi 
đầu vào có phải là anagram của nhau không 

```python
s1 = input()
s2 = input()

dem1 = {}
i = 0
while i < len(s1):
    c = s1[i]
    if c not in dem1:
        dem1[c] = 0
    dem1[c] += 1
    i += 1

dem2 = {}
i = 0
while i < len(s2):
    c = s2[i]
    if c not in dem2:
        dem2[c] = 0
    dem2[c] += 1
    i += 1

if dem1 == dem2:
    print("YES")
else:
    print("NO")
```

## W9A23 
Cho số nguyên rất lớn (string), và một số nguyên dương k. Tìm số có k chữ số lớn nhất có thể 
từ số ban đầu bằng cách xóa bớt đi một số chữ số nào đó mà không thay đổi vị trí của bất kì 
chữ số nào. (VD: 78934340999, k = 7 => 9440999 là số lớn nhất có 7 chữ số cần tìm) 

```python
s = input()
k = int(input())
n = len(s)
xoa = n - k  # số chữ số cần xóa

stack = []
i = 0
while i < len(s):
    while xoa > 0 and len(stack) > 0 and stack[-1] < s[i]:
        stack.pop()
        xoa -= 1
    stack.append(s[i])
    i += 1

# Nếu còn phải xóa nữa, xóa từ cuối
stack = stack[:len(stack) - xoa]
print("".join(stack))
```

## W9A24 
Nén chuối kí tự bằng cách đếm số lần xuất hiện của các kí tự liên tiếp nhau và ghi lại ngay sau 
kí tự đó. VD aaaaaafdhhhhh =>a6f1d1h5 

```python
s = input()
i = 0
result = ""
while i < len(s):
    c = s[i]
    count = 1
    while i + count < len(s) and s[i + count] == c:
        count += 1
    result += c + str(count)
    i += count
print(result)
```

## W9A25 
Giả sử mật khẩu mạnh là một chuỗi kí tự dài ít nhất 8 kí tự gồm ít nhất 1 chữ cái thường, 1 chữ 
cái hoa, 1 chữ số và 1 trong các kí tự đặc biệt !, @, #, $. Nhập vào một chuỗi kĩ tự, kiểm tra xem 
nó có phải là mật khẩu mạnh không. 


```python
s = input()
du_dai = len(s) >= 8
co_thuong = 0
co_hoa = 0
co_so = 0
co_dac_biet = 0
dac_biet = "!@#$"

i = 0
while i < len(s):
    c = s[i]
    if 'a' <= c <= 'z':
        co_thuong = 1
    elif 'A' <= c <= 'Z':
        co_hoa = 1
    elif '0' <= c <= '9':
        co_so = 1
    elif c in dac_biet:
        co_dac_biet = 1
    i += 1

if du_dai and co_thuong and co_hoa and co_so and co_dac_biet:
    print("Strong")
else:
    print("Weak")
```
