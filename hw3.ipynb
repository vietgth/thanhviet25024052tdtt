## Bài 3:
Nhập vào số nguyên dương a, in ra bảng cửu chương của a 

```python
h = int(input())

for i in range(1, h + 1):
    khoang_cach = h - i
    if i == 1:
        print(" " * khoang_cach + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print(" " * khoang_cach + "*" + " " * (2 * i - 3) + "*")
```

## Bài 12:
Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
S = 0
k = 0
while True:
    S = S + (k + 1)
    if S >= n:
        break
    k = k + 1
print("k =", k, ", S(k) =", S - (k + 1))
```

1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

## Bài 14:
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
so_dau = int(input())
if so_dau == -1:
    print("Khong co so nao duoc nhap")
else:
    lon_nhat = so_dau
    nho_nhat = so_dau
    while True:
        x = int(input())
        if x == -1:
            break
        if x > lon_nhat:
            lon_nhat = x
        if x < nho_nhat:
            nho_nhat = x
    print("So lon nhat:", lon_nhat)
    print("So nho nhat:", nho_nhat)
```

## Bài 16:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    n = n // 10
print(chan)
print(le)
```

## Bài 19:
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())
a = 1
b = 1
while b <= A:
    a, b = b, a + b
print(a)
```

## Bài 20:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n = int(input())
def is_prime(x):
    if x < 2:
        return False
    for i in range(2, x):
        if x % i == 0:
            return False
    return True
dem = 0
so = n
while so > 0:
    chu_so = so % 10
    if is_prime(chu_so):
        dem += 1
    so = so // 10
print(dem)
```
