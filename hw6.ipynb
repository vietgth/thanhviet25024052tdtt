## W6A16 
Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải 

```python
n, m = map(int, input().split())
a = []
i = 0
while i < n:
    row = list(map(int, input().split()))
    a.append(row)
    i += 1

i = 0
while i < n:
    j = 0
    while j < m:
        print("%4d" % a[i][j], end="")
        j += 1
    print()
    i += 1
```

## W6A17 
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
n = int(input())
a = []
i = 0
while i < n:
    row = list(map(int, input().split()))
    a.append(row)
    i += 1

print("Cheo chinh:", end="")
i = 0
while i < n:
    print(a[i][i], end=" ")
    i += 1
print()

print("Cheo phu:", end="")
i = 0
while i < n:
    print(a[i][n-1-i], end=" ")
    i += 1
print()
```

## W6A18 
Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên 
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột 
k (hàng k)  

```python
n, m, k = map(int, input().split())
a = []
i = 0
while i < n:
    row = list(map(int, input().split()))
    a.append(row)
    i += 1

tong_cot = 0
i = 0
while i < n:
    tong_cot += a[i][k]
    i += 1

tong_hang = 0
j = 0
while j < m:
    tong_hang += a[k][j]
    j += 1

print("Tong cot", k, "=", tong_cot)
print("Tong hang", k, "=", tong_hang)
```

## W6A19 
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
trong list

```python
def vi_tri_lon_nhat(lst):
    idx = 0
    i = 1
    while i < len(lst):
        if lst[i] > lst[idx]:
            idx = i
        i += 1
    return idx

a = list(map(int, input().split()))
print(vi_tri_lon_nhat(a))
```

## W6A20 
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả 

```python
def tim_vi_tri(lst, k):
    i = 0
    while i < len(lst):
        if lst[i] == k:
            return i
        i += 1
    return -1

a = list(map(int, input().split()))
k = int(input())
print(tim_vi_tri(a, k))
```

## W6A21 
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
words = input().split()
d = {}
i = 0
while i < len(words):
    w = words[i]
    if w not in d:
        d[w] = ()
    d[w] = d[w] + (i,)
    i += 1
print(d)
```

## W6A23 
VIết hàm xóa đi các số <x có trong mảng 

```python
def xoa_nho_hon(lst, x):
    i = 0
    while i < len(lst):
        if lst[i] < x:
            lst.pop(i)
        else:
            i += 1
    return lst

a = list(map(int, input().split()))
x = int(input())
print(xoa_nho_hon(a, x))
```

## W6A24 
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day_0_ve_dau(lst):
    result = []
    zeros = []
    others = []
    i = 0
    while i < len(lst):
        if lst[i] == 0:
            zeros.append(0)
        else:
            others.append(lst[i])
        i += 1
    return zeros + others

a = list(map(int, input().split()))
print(day_0_ve_dau(a))
```

## W6A25 
Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong 
mảng 

```python
a = list(map(int, input().split()))
digit = 0
while digit <= 9:
    found = 0
    i = 0
    while i < len(a):
        if a[i] == digit:
            found = 1
        i += 1
    if found == 0:
        print(digit, end=" ")
    digit += 1
print()
```

## W6A26 
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 
phần tử đầu và cuối 

```python
a = list(map(int, input().split()))
tong = 0
i = 1
while i < len(a) - 1:
    if a[i] > a[i-1] and a[i] > a[i+1]:
        tong += a[i]
    i += 1
print(tong)
```

## W6A27 
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). 
nếu không có in ra -1

## W6A28 
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
thực sự, hay vô danh (không tăng, không giảm) 

```python
a = list(map(int, input().split()))
tang = 1
giam = 1
i = 1
while i < len(a):
    if a[i] <= a[i-1]:
        tang = 0
    if a[i] >= a[i-1]:
        giam = 0
    i += 1

if tang == 1:
    print("Tang thuc su")
elif giam == 1:
    print("Giam thuc su")
else:
    print("Vo danh")
```
