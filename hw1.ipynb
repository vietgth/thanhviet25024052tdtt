## Bài 2: 
Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân. 


```python
a,b=map(int,input().split())
pi = 3.14
r = min(a, b) / 2
S = a * b - pi * r * r
print(f"{S:.2f}")
```

## Bài 3:
Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem
trước cấu trúc lựa chọn: if….else, ma) 

```python
c = input("Nhập một ký tự: ")

if c.isupper():
    print(c.lower())
else:
    print(c.upper())
```

## Bài 4:
Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c}
không phải là kí tự alphabet. 

```python
c = input("Nhập một ký tự: ")

if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

## Bài 5:
Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt.

```python
A = input("Nhập chữ cái hoa: ")
a = A.lower() 
if a == 'a':
    print("Không có chữ cái nào trước 'a'")
else:
    print(chr(ord(a) - 1))  # lấy chữ liền trước
```

## Bài 6:
Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math
a,b,c=map(float,input().split())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = math.sqrt(p * (p-a) * (p-b) * (p-c))
    print(f"Diện tích tam giác: {S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

## Bài 7:
Nhập vào một chuỗi gồm các kí tự thường độ dài >=20, in ra chữ cái thứ năm và
thứ chín của chuỗi này. 

```python
s = input("Nhập chuỗi (độ dài >= 20): ")
if len(s) >= 20:
    print(f"Chữ cái thứ 5: {s[4]}")
    print(f"Chữ cái thứ 9: {s[8]}")
else:
    print("Chuỗi không đủ 20 ký tự!")
```

## Bài 8:
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh

Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng 

```python
ho_ten = input("Nhập họ tên chủ hộ: ")
truoc = int(input("Chỉ số tháng trước: "))
nay   = int(input("Chỉ số tháng này: "))

dien = nay - truoc
tien = 0

if dien <= 50:
    tien = dien * 1984
elif dien <= 100:
    tien = 50*1984 + (dien-50)*2050
elif dien <= 200:
    tien = 50*1984 + 50*2050 + (dien-100)*2380
elif dien <= 300:
    tien = 50*1984 + 50*2050 + 100*2380 + (dien-200)*2998
elif dien <= 400:
    tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 + (dien-300)*3350
else:
    tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 + 100*3350 + (dien-400)*3460

tien_vat = int(tien * 1.08)

print(f"Chủ hộ     : {ho_ten}")
print(f"Số kWh dùng: {dien} kWh")
print(f"Tiền điện  : {tien_vat} đồng")
```
